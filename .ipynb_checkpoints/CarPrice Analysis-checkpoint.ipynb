## Data pre-processing

```python
# feature engineering

df["brand"] = df["CarName"].apply(lambda x : x.split()[0])
```

```python
df["brand"].unique()
```

```python
#fixing brand names

brand_mapping = {'maxda':'mazda',
                 'nissan':'Nissan',
                 'porcshce':'porsche',
                 'toyouta':'toyota',
                 'vokswagen':'volkswagen',
                 'vw':'volkswagen'}
df["brand"] = df["brand"].replace(brand_mapping)
```

```python
df['brand'].unique()
```

```python
df = df.drop(["car_ID", "CarName"], axis=1)
df
```

## Analysis & vizualization

```python
plt.figure(figsize=(14,8))
sns.boxplot(x='brand', y='price', data=df)
plt.xticks(rotation=45)
plt.title("Price distrbution by brand")
plt.show()
```

```python
# Top 5 Brands price

top_brands = df.groupby('brand')['price'].mean().sort_values(ascending = False).head(5)
print(top_brands)
```

```python
# Relationship between engine size and price

plt.figure(figsize=(10,6))
sns.scatterplot(x='enginesize', y='price', data=df, hue='fueltype')
plt.title("Engine size vs price")
plt.show()
```

```python
# Average price by body type

plt.figure(figsize=(10,6))
df.groupby('carbody')['price'].mean().sort_values().plot(kind='bar')
plt.title("Average price by Body Type")
plt.show()
```

```python
# Key numerical features

sns.pairplot(df[["wheelbase","carlength", "carwidth", "curbweight", "enginesize","price"]])
plt.suptitle("pairplot of key Numerical features", y=1.02)
plt.show()
```

```python
# price distrbution by number of cylinders

plt.figure(figsize=(10,6))
sns.boxplot(x="cylindernumber", y='price', data=df)
plt.title('price distrbution by number of cylinders')
plt.show()
```

```python
# horsepower vs price with drive wheel differentiation

plt.figure(figsize=(14,6))
sns.scatterplot(x='horsepower', y='price', data=df, hue='drivewheel', style='aspiration')
plt.title('horsepower vs price with drive wheel differentiation')
plt.show()
```
